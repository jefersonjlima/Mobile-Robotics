```python
import os
os.environ['ROS_MASTER_URI'] = "http://localhost:11311"

import rospy
from sensor_msgs.msg import LaserScan
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Lidar:
    def __init__(self):
        rospy.init_node("lidar_sensor_node", anonymous=True)
        rospy.Subscriber("/scan", LaserScan, self.update)

    def update(self, msg):
        self.scan = msg
```

```python
LIDAR = Lidar()
```

```python
LIDAR.scan.__dir__()
```

```python
len(LIDAR.scan.ranges)
```

```python
plt.plot(LIDAR.scan.ranges)
plt.show()
```

```python
angles = [i for i in range(0, 360)]
angles = np.deg2rad(angles)

ax = plt.subplot(111, projection='polar');
ax.plot(angles, LIDAR.scan.ranges)
```
